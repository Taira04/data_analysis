```python
# ライブラリのインポート
import pandas as pd
import matplotlib.pyplot as plt
import pandas_datareader.data as web
import datetime
from pandas import DataFrame
%matplotlib inline 
```

```python
# データの対象期間設定
start = datetime.datetime(2020, 2, 1)
end = datetime.datetime(2020, 3, 25)
```

```python
# yahoo financeより日経平均株価を取得(dronpa()で欠損値削除)
NIKKEI = web.DataReader("^N225","yahoo",start, end).dropna()#jpy
```

```python
# データフレームの確認
NIKKEI
```

```python
# データ型の確認
NIKKEI.dtypes
```

```python
# 調整後終値(Adj Close)のチャートを描写
NIKKEI.loc[:,'Adj Close'].plot()
```

```python
# 同様にyahoo financeよりナスダック総合指数を取得し、チャートを描写
NASDAQ = web.DataReader("^IXIC","yahoo",start, end).dropna()
NASDAQ.loc[:,'Adj Close'].plot()
```

```python
# 日経・ナスダック共にデータフレームを作成
df_NIKKEI = pd.DataFrame(NIKKEI)
df_NASDAQ = pd.DataFrame(NASDAQ)
```

```python
# 日経・ナスダックのデータフレームに列「指数」を追加しそれぞれに名称を設定
df_NIKKEI['指数'] = '日経平均株価'
df_NASDAQ['指数'] = 'ナスダック総合指数'
```

```python
# 日経とナスダックのデータフレームを縦に結合
df = pd.concat([df_NIKKEI,df_NASDAQ])
```

```python
# データフレームの確認
df
```

```python
# データフレームをカレントディレクトリに「data.csv」というファイル名称でcsv出力
df.to_csv('./date.csv')
```
